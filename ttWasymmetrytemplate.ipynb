```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta = 0.025
x = np.arange(0.0, 10.0, delta)
y = np.arange(0.0, 10.0, delta)
X, Y = np.meshgrid(x, y)
integral = X+Y
acl = (X-Y)/(X+Y)
```

```python
fig, ax = plt.subplots()
acl_c = ax.contour(X, Y, acl)
int_c = ax.contour(X, Y, integral,colors='k')

ax.clabel(acl_c, inline=1, fontsize=10)
ax.clabel(int_c, inline=1, fontsize=10)

ax.set_xlabel("negative bin")
ax.set_ylabel("positive bin")
```
